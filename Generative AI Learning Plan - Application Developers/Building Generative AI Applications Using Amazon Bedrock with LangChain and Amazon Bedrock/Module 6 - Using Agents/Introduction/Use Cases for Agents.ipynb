# **Use Cases for Agents**
- **Amazon Bedrock Agents** enable **LLMs** to **run tasks, interact with humans**, and **retrieve accurate, up-to-date information**.
- They **integrate domain knowledge, contextual understanding, and task execution** to **automate workflows**.

- **Knowledge integration** - Agents can integrate external knowledge sources.
- **Context-aware responses** - Agents maintain short-term conversational context.
- **Complex task execution** - Agents can break down complex tasks into smaller subtasks and distribute them to appropriate models.
- **Transparency** - Agents can share the reasoning for the decisions and the outputs generated by LLMs.
- **Reliability and credibility** - Agents can act as another layer to interpret LLM outputs.
- **Personalization and adaptation** - Agents can adapt to individual user preferences, learning patterns, and requirements.
- **Task automation and workflow optimization** - Agents can automate repetitive tasks and optimize workflows with the capabilities of LLMs.
