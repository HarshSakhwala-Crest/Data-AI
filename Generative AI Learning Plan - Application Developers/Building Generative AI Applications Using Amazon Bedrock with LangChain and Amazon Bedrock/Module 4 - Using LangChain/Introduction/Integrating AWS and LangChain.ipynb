# Integrating AWS and LangChain

## **Supported Integrations for AWS**
LangChain supports integrations with various **AWS services and AI models**. These integrations enable functionalities such as:
- **Chat models**
- **Large Language Models (LLMs)**
- **Text embedding models**
- **Document loaders**
- **Vector stores**
- **Retrievers**
- **Memory management**
- **Graphs**
- **Callbacks**
- **Chains and Agents**

### *Amazon Bedrock Integrations**
LangChain integrates with **Amazon Bedrock** using:
- **`langchain.aws`** package
- **`langchain-community`** package

## **Key Takeaways**
- **LangChain seamlessly integrates with AWS**, supporting various **LLMs, vector stores, and memory management solutions**.
- **Amazon Bedrock supports multiple FMs**, including **Titan, Jurassic, Claude, Cohere, Llama, Stable Diffusion, and Mistral**.
- **AWS LangChain integrations** provide a structured way to manage **LLMs, prompts, memory, and multi-step workflows**.
