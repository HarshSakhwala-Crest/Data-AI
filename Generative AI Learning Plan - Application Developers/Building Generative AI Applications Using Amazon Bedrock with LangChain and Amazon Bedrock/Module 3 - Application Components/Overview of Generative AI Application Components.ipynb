# Overview of Generative AI Application Components

## Typical components of a generative AI application

The following diagram includes the typical components involved in building a generative artificial intelligence (generative AI) application.

A foundation model (FM) is in the center surrounded by components. These include a frontend web application or mobile app, FM interface, machine learning (ML) environment, model training, enterprise datasets, vector database, text and image embeddings, and long-term memory store. Governance and security are integrated into all components.
